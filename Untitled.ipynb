# Time calibration
Measuring the delays in the signals between two plates in one configuration and then flipping them and repeating the measurement allows for the removal of the delay due to the cables.

In the first configuration we find:

$-4.97815e+01 \pm  1.59768e-02 = t_1-t_0$

$-2.63188e+01 \pm  2.22969e-02 = t_2-t_0$

$-2.68886e+01 \pm  2.45819e-02 = t_3-t_0$

Flipping the detector:

$-5.01835e+01 \pm  1.06227e-04 = t_1-t_0$

$-2.81133e+01 \pm  1.36318e-04 = t_2-t_0$

$-2.90723e+01 \pm  5.22752e-04 = t_3-t_0$
